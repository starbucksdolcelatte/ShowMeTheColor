# Concat

```python
cheek = pd.concat([cheek_colors_0, cheek_colors_1, cheek_colors_2, cheek_colors_3])
cheek.index = cheek.label
cheek['R'].groupby(cheek['label']).size()
```

```python
from scipy import stats
cheek[(np.abs(stats.zscore(cheek.H)) < 3)]
```

# Scatter

```python
# Scatter plot with a different color by groups

groups = cheek.groupby('label')
fig, ax = plt.subplots(3, 3, figsize = (10,20))

for name, group in groups:
    ax[0][0].plot(
            group.R[abs(stats.zscore(group.R))<1.5],
            marker='o', 
            linestyle='',
            label=name)
for name, group in groups:
    ax[0][1].plot(
            group.G[abs(stats.zscore(group.G))<1.5],
            marker='o', 
            linestyle='',
            label=name)
for name, group in groups:
    ax[0][2].plot(
            group.B[abs(stats.zscore(group.B))<1.5], 
            marker='o', 
            linestyle='',
            label=name)
    
for name, group in groups:
    ax[1][0].plot(
            group.Lab_L[abs(stats.zscore(group.Lab_L))<1.5], 
            marker='o', 
            linestyle='',
            label=name)
for name, group in groups:
    ax[1][1].plot(
            group.Lab_a[abs(stats.zscore(group.Lab_a))<1.5], 
            marker='o', 
            linestyle='',
            label=name)
for name, group in groups:
    ax[1][2].plot(
            group.Lab_b[abs(stats.zscore(group.Lab_b))<1.5], 
            marker='o', 
            linestyle='',
            label=name)
    
for name, group in groups:
    ax[2][0].plot(
            group.H[abs(stats.zscore(group.H))<1.5], 
            marker='o', 
            linestyle='',
            label=name)
for name, group in groups:
    ax[2][1].plot(
            group.S[abs(stats.zscore(group.S))<1.5], 
            marker='o', 
            linestyle='',
            label=name)
for name, group in groups:
    ax[2][2].plot(
            group.V[abs(stats.zscore(group.V))<1.5], 
            marker='o', 
            linestyle='',
            label=name)

#ax.legend(fontsize=12, loc='upper left') # legend position
plt.title('Scatter Plot of iris by matplotlib', fontsize=20)
plt.show()
```
